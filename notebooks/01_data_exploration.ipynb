```python
import pandas as pd

df = pd.read_csv("../data/processed/season_2025_features.csv")

team_baseline = (df.groupby("team").agg(
    avg_points=("points", "mean"),
    avg_finish=("finish_position", "mean"),
    wins=("is_win", "sum")
    )
    .reset_index()
)

team_baseline
```

```python
team_baseline["team_base_rating"] = (
    0.5 * team_baseline["avg_points"] +
    0.3 * (1 / team_baseline["avg_finish"]) +
    0.2 * team_baseline["wins"]
)
team_baseline.head(10)
```

```python
team_baseline["team_base_rating"] = (
    0.5 * team_baseline["avg_points"] +
    0.3 * (1 / team_baseline["avg_finish"]) +
    0.2 * team_baseline["wins"]
)

# Normalize supaya skala lebih masuk akal
team_baseline["team_base_rating"] = (
    (team_baseline["team_base_rating"] - team_baseline["team_base_rating"].mean())
    / team_baseline["team_base_rating"].std()
)

team_baseline.sort_values("team_base_rating", ascending=False)
```
